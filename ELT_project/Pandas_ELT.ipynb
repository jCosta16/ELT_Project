```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
fifa_transfermark = "data/fifa_transfermark_df.csv"
fifa_transfermark_df = pd.read_csv(fifa_transfermark)
fifa_transfermark_df = fifa_transfermark_df.reset_index()
```

```python
fifa_transfermark_df
```

# Create a Database Conection

```python
new_fifa_transfermark_df = fifa_transfermark_df[['index','Name','age','club','overall','potential','value_eur','nationality','Position','Market Value']].copy()
new_fifa_transfermark_df = new_fifa_transfermark_df.rename(columns= {'index':'id', 'Name':'name', 'Position': 'position',\
                                                                     'value_eur': 'game_value','Market Value':'market_value' })
# new_premise_df.drop_duplicates("id", inplace=True)
new_fifa_transfermark_df
```

```python
rds_connection_string = "postgres:PostgreSQL@localhost:5432/fifa_transfermarkt_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

# Load the Dataframe into the Database

```python
new_fifa_transfermark_df.to_sql(name='fifa_transfermarkt', con=engine, if_exists='append', index=False)
```
